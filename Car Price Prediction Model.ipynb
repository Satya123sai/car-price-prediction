```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
x_tr,x_ts,y_tr,y_ts=train_test_split(x,y,test_size=0.15)
```

```python
ohe =OneHotEncoder()
```

```python
ohe.fit(df[['company','transmission','fuel','seller_type','owner']])
```

```python
ct=make_column_transformer((OneHotEncoder(),['company','transmission','fuel','seller_type','owner']),remainder="passthrough")
```

```python
lr=LinearRegression()
```

```python
model=make_pipeline(ct,lr)
```

```python
model.fit(x,y)
```

```python
pred=model.predict(x_ts)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_ts,pred)
```
